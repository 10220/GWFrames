I've implemented a new fourth-order scalar derivative function in `GWFrames`, and would like to check that the results are reasonable.  The same code (just substituting data types) is also used for complex derivatives, and in the `Quaternions` module, so checking the scalar derivative should be sufficient for checking each of those.

The result is that we indeed have beautiful fourth-order convergence.  I even found a couple bugs by noticing that the ends were not symmetric and did not do exactly what I expected.

```python
%pylab inline
%config InlineBackend.figure_format='svg'

import GWFrames
```

```python
x1 = linspace(0, 2*pi, num=1000)
x2 = x1[::2]
x4 = x1[::4]
x8 = x1[::8]

y1 = sin(x1)
y2 = sin(x2)
y4 = sin(x4)
y8 = sin(x8)

yp1 = cos(x1)
yp2 = cos(x2)
yp4 = cos(x4)
yp8 = cos(x8)

dydx1 = GWFrames.ScalarDerivative(y1, x1)
dydx2 = GWFrames.ScalarDerivative(y2, x2)
dydx4 = GWFrames.ScalarDerivative(y4, x4)
dydx8 = GWFrames.ScalarDerivative(y8, x8)
```

```python
semilogy(x1, abs(yp1-dydx1));
semilogy(x2, abs(yp2-dydx2));
semilogy(x4, abs(yp4-dydx4));
semilogy(x8, abs(yp8-dydx8), '.-');
```

```python
semilogy(x1, abs(yp1-dydx1));
semilogy(x2, abs(yp2-dydx2)/2.**4);
semilogy(x4, abs(yp4-dydx4)/4.**4);
semilogy(x8, abs(yp8-dydx8)/8.**4, '.-');
```
