# Direct use of `TransformToCorotatingFrame`

It would be nice to be able to take a `PNWaveform` (which is constructed in the co-orbital frame) and transform directly to the co-rotating frame without first transforming to the inertial frame.  This will be faster and more accurate.  I've changed `TransformToCorotatingFrame` to be able to do this, so I need to test that the old method gives the same results as the new method (to within numerical accuracy).

First, set up a `PNWaveform` with some basic precession

```python
Approximant = 'TaylorT4' # 'TaylorT1'|'TaylorT4'|'TaylorT5'
delta = 0.1 # Normalized BH mass difference (M1-M2)/(M1+M2)
chi1_i = [0.1,0.2,0.3] # Initial dimensionless spin vector of BH1
chi2_i = [0.5,0.2,0.1] # Initial dimensionless spin vector of BH2
Omega_orb_i = 0.014 # Initial orbital angular frequency

W1 = GWFrames.PNWaveform(Approximant, delta, chi1_i, chi2_i, Omega_orb_i)
```

Now, we want to test whether we get the same thing by first transforming to the inertial frame.  So we make a copy of the first `PNWaveform` and transform it to the inertial frame.

```python
W2 = GWFrames.PNWaveform(W1)
W2.TransformToInertialFrame();
```

```python
AV1 = W1.AngularVelocityVectorRelativeToInertial([2,3,4])
AV2 = W2.AngularVelocityVectorRelativeToInertial([2,3,4])
display(AV1)
display(AV2)
semilogy(W1.T(), norm(AV1-AV2, axis=1));
```

We see that their `AngularVelocityVectorRelativeToInertial`s are pretty comparable, throughout most of the evolution, with differences of $10^{-9}$ that look to me like they're consistent with time-stepping or differentiation error.  The differences blow up in the last ${\sim}10M$ or so, but this is not unusual for things dealing with PN.

```python
AV3 = W1.AngularVelocityVector([2,3,4])
AV4 = W2.AngularVelocityVector([2,3,4])
display(AV3)
display(AV4)
semilogy(W1.T(), norm(AV3, axis=1), label='W1');
semilogy(W1.T(), norm(AV4, axis=1), label='W2');
legend(loc='upper left');
```

We can see that the `AngularVelocityVector`s relative to the `PNWaveform`s' respective frames are quite different, which is what we expect.  But each one individually is fairly large, especially compared to the differences we saw above.  So this suggests that our various measurements make sense and are consistent.

Next, we do the actual transformations into the co-rotating frame, and check that the results are consistent, regardless of whether we first transform to the inertial frame.

```python
W3 = GWFrames.PNWaveform(W1)
W3.TransformToCorotatingFrame([2,3,4])
W4 = GWFrames.PNWaveform(W2)
W4.TransformToCorotatingFrame([2,3,4]);
display(W3.Frame())
display(W4.Frame())
semilogy(W1.T(), Quaternions.angle(Quaternions.RDelta(W3.Frame(), W4.Frame())))
#xlim(xmin=2450)
```

```python
semilogy(W3.T(), abs(W3.Arg(W3.FindModeIndex(2,2))-W4.Arg(W4.FindModeIndex(2,2))), lw=5, label=r'$2,2$')
semilogy(W3.T(), abs(W3.Arg(W3.FindModeIndex(2,-2))-W4.Arg(W4.FindModeIndex(2,-2))), lw=2, label=r'$2,-2$')
legend(loc='upper left');
```

The differences between the frame rotors and the waveform mode phases themselves for the two methods are again generally small -- and entirely consistent with the differences between the angular velocities seen above.

So, we conclude that `GWFrames` can now successfully transform directly from the co-orbital frame to the co-rotating frame.

# `EvaluateAtPoint`

Now, we need to check that my updates to `EvaluateAtPoint` -- allowing it to operate on `Waveform`s in rotating frames -- didn't change the behavior significantly.  First, we check that I haven't changed the behavior for `Waveform`s in the inertial frame.

```python
vartheta, varphi = 0.12, 0.34
```

```python
# EvaluateAtPoint1 is the "classic" version, which will be removed momentarily
inertial_diff = W2.EvaluateAtPoint(vartheta, varphi)-W2.EvaluateAtPoint1(vartheta, varphi)
max(abs(inertial_diff))
```

They are identical, as I had hoped.  Now, we can compare the results from `W1` and `W2`, which should only be the same up to the accuracies we saw above.

```python
d1 = W1.EvaluateAtPoint(vartheta, varphi);
d2 = W2.EvaluateAtPoint(vartheta, varphi);
```

```python
plot(W1.T(), d1.real);
plot(W1.T(), d1.imag);
plot(W2.T(), d2.real);
plot(W2.T(), d2.imag);
ylim(-0.1, 0.1)
xlim(0,3000);
```

```python
semilogy(W1.T(), abs(d1-d2));
semilogy(W1.T(), abs(d1-d2)/abs(d1));
```

These are even more identical than I expected, with differences essentially at roundoff.

Let's just evaluate the max of that plot (except the last 20 values, let's say) for various special and random positions.

```python
max_abs = 0.0
for vartheta, varphi in [[0.,0.], [0.,np.pi], [0.,2*np.pi], [np.pi,np.pi]]+rand(50,2).tolist():
    max_abs = max(max_abs, max(abs(W1.EvaluateAtPoint(vartheta, varphi)-W2.EvaluateAtPoint(vartheta, varphi))[:-20]))
print(max_abs)
```

There are $N_{\text{times}}\, (2\ell_{\text{max}}+1)^2$ interpolations, so if the interpolators need to be allocated each time, that can be a huge overhead:

```python
W1.NTimes()*(2*W1.EllMax()+1)**2
```

For the current `Waveform`, that's fourteen million function calls, with four allocations each time.  Let's see how the current function treats this:

```python
%time W2.Translate([[0.001*t, 0.002*t, 0.006*t] for t in W6.T()]);
```
