```python
import GWFrames
import GWFrames.plot
import Quaternions
```

```python
Approximant = 'TaylorT4' # 'TaylorT1'|'TaylorT4'|'TaylorT5'
delta = 0.1 # Normalized BH mass difference (M1-M2)/(M1+M2)
chi1_i = [0.1,0.2,0.3] # Initial dimensionless spin vector of BH1
chi2_i = [0.5,0.2,0.1] # Initial dimensionless spin vector of BH2
Omega_orb_i = 0.014 # Initial orbital angular frequency

W1 = GWFrames.PNWaveform(Approximant, delta, chi1_i, chi2_i, Omega_orb_i)
```

```python
W2 = GWFrames.PNWaveform(W1)
W2.TransformToInertialFrame();
W1.TransformToCorotatingFrame();
```

```python
%%time
W5 = GWFrames.Waveform(W1.Translate([[0.001*t, 0.002*t, 0.006*t] for t in W1.T()]));
```

```python
%%time
W6 = GWFrames.Waveform(W2.Translate([[0.001*t, 0.002*t, 0.006*t] for t in W2.T()]));
```

```python
W7 = GWFrames.Waveform(W6)
W7.TransformToCorotatingFrame();
```

```python
W5.AlignDecompositionFrameToModes(2000.0);
W7.AlignDecompositionFrameToModes(2000.0);
# GWFrames.AlignWaveforms(W5, W7, 1000, 3000);
```

```python
W5.plot('Abs', [[2,2],[2,-2]], label=5)
W7.plot('Abs', [[2,2],[2,-2]], label=7)
legend();
```
