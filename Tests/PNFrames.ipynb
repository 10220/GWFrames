```python
import GWFrames
import GWFrames.plot
```

```python
PN = GWFrames.PNWaveform('TaylorT4', 0.01, [0.4, 0.0, 0.1], [0.0, 0.6, 0.1], 0.01)
```

```python
PN.NTimes()
```

```python
PN.plot('LogAbs', [2,2])
```

```python
PN.FrameTypeString()
```

```python
PN_inertial = GWFrames.PNWaveform(PN)
PN_inertial.TransformToInertialFrame();
PN_corot1 = GWFrames.PNWaveform(PN_inertial)
PN_corot1.TransformToCorotatingFrame();
PN_corot2 = GWFrames.PNWaveform(PN)
PN_corot2.TransformToCorotatingFrame();
```

```python
PN.plot('LogAbs', [2,2])
PN_inertial.plot('LogAbs', [2,2])
PN_corot1.plot('LogAbs', [2,2])
PN_corot2.plot('LogAbs', [2,2])
```

```python
PN_inertial.plot('LogAbs', [[2,2]], label=0)
PN_corot1.plot('LogAbs', [[2,2]], label=1)
PN_corot2.plot('LogAbs', [[2,2]], label=2)
ylim((7e-2,1e-1));
legend();
```
