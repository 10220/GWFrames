```python
import sys
import numpy as np
import scipy.optimize
import h5py
import matplotlib as mpl
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D # For 3D plots
#mpl.rcParams['figure.figsize'] = 8,6  # Set default figure size (in inches)
#%config InlineBackend.figure_format='svg'  # If plotting inline, use SVG

import GWFrames
import GWFrames.plot
import Quaternions

PathToSXSCatalog = '/Users/boyle/Research/Data/SimulationAnnex/Catalog/'
MyFavoriteSystem = PathToSXSCatalog + 'q1.0/SBBH/d19.0_q1.0_s0.5_0_0_s0_0_0/Lev5/'
MyFavoriteWaveform = MyFavoriteSystem + 'rhOverM_Asymptotic_GeometricUnits.h5/Extrapolated_N2.dir'

W_NR = GWFrames.ReadFromNRAR(MyFavoriteWaveform);
t0 = -W_NR.MaxNormTime()
W_NR.SetT(W_NR.T()+t0)
print('The initial time has been moved back to {0}'.format(t0))

# Get the data from the h5 file
with h5py.File(MyFavoriteSystem+'Horizons.h5', 'r') as f:
    tA = f['AhA.dir/CoordCenterInertial.dat'][:,0]+t0
    xA = f['AhA.dir/CoordCenterInertial.dat'][:,1:]
    mA = f['AhA.dir/ArealMass.dat'][:,1]
    chiA = f['AhA.dir/chiInertial.dat'][:,1:]
    tB = f['AhB.dir/CoordCenterInertial.dat'][:,0]+t0
    xB = f['AhB.dir/CoordCenterInertial.dat'][:,1:]
    mB = f['AhB.dir/ArealMass.dat'][:,1]
    chiB = f['AhB.dir/chiInertial.dat'][:,1:]

# Some calculations with the coordinates
d = xA-xB
# nHat = array([[q[1],q[2],q[3]] for q in Quaternions.normalized(Quaternions.QuaternionArray(d))])
# dnHatdt = array([[q[1],q[2],q[3]] for q in Quaternions.QuaternionDerivative(Quaternions.QuaternionArray(nHat), tA)])
# lambdaHat = array([[q[1],q[2],q[3]] for q in Quaternions.normalized(Quaternions.QuaternionArray(dnHatdt))])
nHat = Quaternions.normalized(Quaternions.QuaternionArray(d))
dnHatdt = Quaternions.QuaternionDerivative(nHat, tA)
lambdaHat = Quaternions.normalized(dnHatdt)
Omega = array([cross(n.vec(),ndot.vec()) for n,ndot in zip(nHat,dnHatdt)])

# Take initial data to be simply the coordinate quantities at the relaxation time
t_relaxed = ! awk {'"/^relaxed-measurement-time/ {print \$3}"'} {MyFavoriteSystem}/metadata.txt
t_relaxed = float(t_relaxed[0])+t0
i_relaxed = abs(tA-t_relaxed).argmin()
ma = mA[i_relaxed]
mb = mB[i_relaxed]
delta = (ma-mb)/(ma+mb)
chia_0 = chiA[i_relaxed]
chib_0 = chiB[i_relaxed]
Omega_orb_0 = Omega[i_relaxed,2]
R_frame_i = Quaternions.FrameFromXY([nHat[i_relaxed],],[lambdaHat[i_relaxed],])[0]

# Alignment and hybridization range:
t1, t2 = -10000., -6000.

# UseEllModes = [2,3,4] # Which ell modes to use to calculate the angular velocity and hence corotating frame
# t_alignment = t0+2000. # Time at which to align the modes to the frame
# W_NR_corot = GWFrames.Waveform(W_NR)
# W_NR_corot.TransformToCorotatingFrame(UseEllModes)
# #%timeit W_NR_corot.AlignDecompositionFrameToModes(t_alignment, UseEllModes);
```

```python
W_NR.plot('LogAbs', [2,2])
```

```python
# Construct complete waveform
W_PN = GWFrames.PNWaveform("TaylorT1", delta, chia_0, chib_0, Omega_orb_0, 0.5*Omega_orb_0, R_frame_i)
W_PN.TransformToInertialFrame();
```

```python
W_PN.plot('LogAbs', [2,2])
W_NR.plot('LogAbs', [2,2])
ylim((7e-2,5e-1))
xlim((W_NR.T(0), W_NR.T()[-1]));
```

```python
W_PN_corot = GWFrames.Waveform(W_PN)
W_NR_corot = GWFrames.Waveform(W_NR)

W_PN_corot.TransformToCorotatingFrame()
W_NR_corot.TransformToCorotatingFrame()

W_PN_corot.plot('LogAbs', [2,2])
W_NR_corot.plot('LogAbs', [2,2])
ylim((4e-2,5e-1))
xlim((W_NR_corot.T(0), W_NR_corot.T()[-1]));
```

The dissimilarity between NR and PN in the plot above show clearly that though they are both in "the" co-rotating frame, that is not enough to uniquely specify the frame; we must remember to align the frame to the modes.  That is not done by default, because there is no good unique and general choice of frame.  But if we select a time at which both should align to their modes, and choose some dumb default $X$ axis, it makes sense:

```python
W_PN_corot.AlignDecompositionFrameToModes((t1+t2)/2.)
W_NR_corot.AlignDecompositionFrameToModes((t1+t2)/2.)

W_PN_corot.plot('LogAbs', [2,2])
W_NR_corot.plot('LogAbs', [2,2])
ylim((7e-2,5e-1))
xlim((W_NR_corot.T(0), W_NR_corot.T()[-1]));
```

```python
W_hyb_corot = GWFrames.Waveform(W_PN_corot.Hybridize(W_NR_corot, -10000., -6000.))
```
