# Sec. IV: The corotating frame

### Sec. IV A: Finding the corotating frame

Given an array of quaternions representing the angular-velocity vector as a function of time, we can derive the appropriate corotating frame.  [Note that Sec. IV B demonstrates a function that computes this frame and transforms the `Waveform` in one easy step.]

```python
# Measure the angular-velocity vector from the waveform
AV_PN = array(W_PN.AngularVelocityVector())

# Get an array of quaternions representing the corotating frame
R_corot = Quaternions.FrameFromAngularVelocity(Quaternions.Quaternions(AV_PN), W_PN.T())
```

The system starts out mostly rotating about the $z$ axis, due to the orbital motion, but also precessing a little.  The frame at the third time step has been rotated about an axis that is mostly parallel to $z$, but has a slight $x$ component:

```python
Rotor = R_corot[2]
Angle = Rotor.angle()
Axis = array(Rotor.log().vec())*(2/Angle)
Time = W_PN.T(2)-W_PN.T(0)
print("Time  = {t}M after beginning\nRotor = {r}\nAxis  = {ax}\nAngle = {a}".
       format(r=Rotor, ax=Axis.tolist(), a=Angle, t=Time))

print("""\nNote that we started the system at an angular frequency of {O}, and the system
has gone through an angle {a:.4} after time {t:.4}, which makes sense.""".format(O=Omega0, a=Angle, t=Time))
```

### Sec. IV B: Gravitational waveforms in the corotating frame

Now, rather than explicitly asking for the angular velocity, then the corotating frame, and then transforming the waveform, we can do it all in one go with the simple `TransformTo` methods:

```python
# Make copies to transform.  Because Python makes copies by reference,
# to make a real copy, we need to use the copy constructor:
W_PN_Corot = GWFrames.PNWaveform(W_PN)
W_PN_Aligned = GWFrames.PNWaveform(W_PN)

# Now, transform the copies to the corotating frame and aligned frame
W_PN_Corot.TransformToCorotatingFrame()
W_PN_Aligned.TransformToOShaughnessyEtAlFrame()

# And plot the waveforms in the different frames
ell=2
ModeIndices = [W_PN.FindModeIndex(ell,m) for m in range(-ell,ell+1)]
plt.figure(1)
plt.subplot(321)
for ModeIndex in ModeIndices : plt.plot(W_PN.T(), W_PN.Abs(ModeIndex))
ylabel(r"""$|r\, h^{2,m}/M\, |$""");
plt.subplot(322)
for ModeIndex in ModeIndices : plt.plot(W_PN.T(), W_PN.ArgUnwrapped(ModeIndex))
ylabel(r"""$\phi^{2,m}$ (radians)""");
plt.subplot(323)
for ModeIndex in ModeIndices : plt.plot(W_PN_Aligned.T(), W_PN_Aligned.Abs(ModeIndex))
ylabel(r"""$|r\, h^{2,m}/M\, |$""");
plt.subplot(324)
for ModeIndex in ModeIndices : plt.plot(W_PN_Aligned.T(), W_PN_Aligned.ArgUnwrapped(ModeIndex))
ylabel(r"""$\phi^{2,m}$ (radians)""");
plt.subplot(325)
for ModeIndex in ModeIndices : plt.plot(W_PN_Corot.T(), W_PN_Corot.Abs(ModeIndex))
ylabel(r"""$|r\, h^{2,m}/M\, |$""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
plt.subplot(326)
for ModeIndex in ModeIndices : plt.plot(W_PN_Corot.T(), W_PN_Corot.ArgUnwrapped(ModeIndex))
ylabel(r"""$\phi^{2,m}$ (radians)""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
```

### Sec. IV C: Comparison, alignment, and hybridization

We can also compare, align, and hybridize any two `Waveform` objects.  (Extrapolation is beyond the scope of this documentation, though several functions provided here can certainly be useful.)  First, we construct two objects that should be very nearly the same.

```python
# Create the waveforms in their inertial frames
W_A = GWFrames.PNWaveform(W_PN);
W_B = GWFrames.PNWaveform((m1-m2)/(m1+m2), chi1, chi2, 0.999*Omega0);
W_B.TransformToInertialFrame();
```

Note that the spin directions are defined at two slightly different frequencies, so these waveforms do not describe the same physics.  But we choose to compare them just so there is some nontrivial difference for demonstration purposes.  We can construct a new `Waveform` object, which records the difference between the modes and the frames.  In this case, the frames are the same (because they are inertial), but the modes are very different:

```python
W_diff = W_A.Compare(W_B)
figure()
for m in range(-2,2+1) : plot(W_diff.T(), W_diff.Re(W_diff.FindModeIndex(2,m)), label=r'$m={0}$'.format(m));
ylabel(r"""$\Delta \Re\{r\, h^{2,m}/M\, \}$""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
legend();
```

Now, we can also align the two waveforms.  Because these systems are highly precessing, their angular frequencies are not monotonic -- and in particular, are not invertible everywhere.  But we are hoping to align them by comparing angular frequencies, so we have to find a suitable time for which the angular frequency is invertible.  We will be leaving `W_B` fixed, so we look for an angular frequency in that waveform that only occurs at one time.  We plot the quantity below, and find that $(t-t_{\mathrm{merger}})/M = -3075,$ &nbsp; for example, is a suitable value.

```python
W_A_AVMag = [GWFrames.abs(a) for a in W_A.AngularVelocityVector()]
W_B_AVMag = [GWFrames.abs(a) for a in W_B.AngularVelocityVector()]
figure()
semilogy(W_A.T()[10:], W_A_AVMag[10:], label="W_A");
semilogy(W_B.T()[10:], W_B_AVMag[10:], label="W_B");
ylabel(r"""$|\, \vec{\omega}\, |$""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
legend();
```

Then, we can align the two Waveforms by ensuring that they have the same $|\vec{\omega}|$ at that moment:

```python
# Do the alignment
t_fid = -3075
W_A.AlignTime(W_B, t_fid);

# Now, plot the angular velocities, zooming in near -3075
W_A_AVMag = [GWFrames.abs(a) for a in W_A.AngularVelocityVector()]
figure()
plot(W_B.T()[10:], W_B_AVMag[10:], label="W_B");
plot(W_A.T()[10:], W_A_AVMag[10:], label="W_A");
xlim([-3500,-2500]);
ylim([2e-2,2.3e-2]);
ylabel(r"""$|\, \vec{\omega}\, |$""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
legend();
```

Note that the curves above intersect at -3075.

However, this only aligns the times at that instant.  We may also want to align the Waveforms.  To do this, we move into a physically meaningful frame (the corotating frame), ensuring that the modes are truly aligned to that frame, and then ensure that the frames are the same:

```python
# Transform to corotating frames
W_A.TransformToCorotatingFrame();
W_B.TransformToCorotatingFrame();

# Ensure that the modes and frames in each Waveform are aligned to each other
W_A.AlignDecompositionFrameToModes(t_fid)
W_B.AlignDecompositionFrameToModes(t_fid)

# Now simply align the frames to each other
W_A.AlignFrame(W_B, t_fid);

# Show the differences
W_diff = W_A.Compare(W_B)
figure()
for m in range(-2,2+1) : plot(W_diff.T(), W_diff.Re(W_diff.FindModeIndex(2,m)), label=r'$m={0}$'.format(m));
ylabel(r"""$\Delta \Re\{r\, h^{2,m}/M\, \}$""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
legend();
figure()
plot(W_diff.T(), [r.angle() for r in W_diff.Frame()]);
ylabel(r"""$\Phi_{\Delta}$ (radians)""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
```

Here, we see that the frames are exactly equal at -3075, and the modes are much more nearly equal than they were initially.

On the other hand, aligning at one instant may not be satisfactory.  Comparing the plots of $|\vec{\omega}|$ above, we see that the secular angular frequencies don't seem to match, even though the instantaneous ones do.  We might want to "smooth out" the cyclical changes by minimizing the difference over a region.  In particular, we add a constant time offset $\delta t$ and a constant rotation offset $R_\delta$ to minimize

$ \int_{t_1}^{t_2} 4 \left| {R_{\mathrm{A}}}(t)\\, {\bar{R}\_{\mathrm{B}}} (t+\delta t) \\, \bar{R}\_\\delta \right|^2\\, dt $

Choosing $t_1$ and $t_2$ to lie around -3075, for example:

```python
# Align over a region
W_A.AlignTimeAndFrame(W_B, -3500, -2500);

# Show the differences
W_diff = W_A.Compare(W_B)
figure()
for m in range(-2,2+1) : plot(W_diff.T(), W_diff.Re(W_diff.FindModeIndex(2,m)), label=r'$m={0}$'.format(m));
ylabel(r"""$\Delta \Re\{r\, h^{2,m}/M\, \}$""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
legend();
figure()
plot(W_diff.T(), [r.angle() for r in W_diff.Frame()]);
ylabel(r"""$\Phi_{\Delta}$ (radians)""");
xlabel(r"$(t-t_{\mathrm{merger}})/M$");
```

We see that the mode differences are essentially the same, but the frames are aligned more closely over a broader range of time.

# Appendix A: Quaternions and rotations

### Appendix A 1: Elements of quaternion mathematics

We create a Quaternion with components (1,2,3,5) as:

```python
Q = Quaternions.Quaternion(1,2,3,5) # These components are just randomly made up as an example
```

Its conjugate is

```python
Q.conjugate()
```

We can multiply by another quaternion:

```python
Q2 = Quaternions.Quaternion(7,11,13,17) # Also randomly made-up components
Q*Q2
```

Notice that the product is *not* commutative in general:

```python
Q2*Q  # This result is not the same as the last result
```

The norm of a quaternion is given by

```python
sqrt( Q * Q.conjugate() )
```

or, equivalently

```python
Q.abs()
```

The inverse of a quaternion is given by

```python
Qinv = Q.conjugate() / Q.normsquared()
Q * Qinv # We might get some roundoff garbage, but this is basically 1
```

We can normalize a quaternion and obtain unit quaternion (which is a rotor):

```python
R = Q.normalized()
R.abs()
```

Ordinary vectors are just quaternions with 0 scalar part:

```python
x = Quaternions.Quaternion(0,1,0,0)
y = Quaternions.Quaternion(0,0,1,0)
z = Quaternions.Quaternion(0,0,0,1)
```

This allows us to define a rotor by its axis-angle description.  The following uses the quaternion exponential function to give us a rotation about the $x$ axis through the angle $\pi/2$:

```python
theta = pi/2.
R = Quaternions.exp(theta * x / 2)
R
```

So, for example, we can rotate $y$ to get $z$ (within numerical precision):

```python
R * y * R.conjugate()
```

Or $z$ to get $-y$:

```python
R * z * R.conjugate()
```

We also have a convenient way to calculate the logarithm of a quaternion.  For our rotor above, we get

```python
R.log()
```

In this case, we get the correct inverse.  However, note that in general we have branch-cut discontinuities as in the case of the usual complex logarithm.  For example, the following exponential gives the same rotor as $R$ above:

```python
Quaternions.exp(theta*x/2 + 2*pi*x)
```

### Appendix A 2: Formulas for rotations of SWSHs

Though they may not be directly useful for our purposes, this module includes implementations of the Wigner $\mathfrak{D}$ matrices and the spin-weighted spherical harmonics in terms of quaternions directly, with no transformations to Euler angles or other representations.

```python
R = Quaternions.exp(pi * Quaternions.Quaternion(0,1,0,0) / 4) # Rotate by pi/2 about x
D = SphericalFunctions.WignerDMatrix()
D.SetRotation(R)
D(2,1,-1) # Evaluate the ell=2, m'=1, m=-1 component of the Wigner D matrix with the rotation set above
```

```python
s = -2 # The spin weight
Ra = Quaternions.Quaternion(pi/2,pi/2) # Rotor corresponding to (vartheta, varphi) = (pi/2,pi/2)
Ylm = SphericalFunctions.SWSH(s, Ra)
[Ylm(2,m) for m in [-2,-1,0,-1,2]]
```

It is very important to check that composition of rotations is equivalent to matrix multiplication of the Wigner $\mathfrak{D}$ matrices, which is used in Appendix B for deriving the SWSH rotation law, and hence the formula for rotating modes of a spin-weighted field.

```python
# Create two random rotors
R1 = Quaternions.normalized(Quaternions.Quaternion(1,2,3,4))
R2 = Quaternions.normalized(Quaternions.Quaternion(8,7,6,5))

# Construct the corresponding D matrices
D1 = SphericalFunctions.WignerDMatrix(R1)
D2 = SphericalFunctions.WignerDMatrix(R2)
D3 = SphericalFunctions.WignerDMatrix(R1*R2)

# Check that D3 = D1*D2
Tolerance = 3e-15
Differences = [abs( D3(ell,mp,m) - sum([D1(ell,mp,mpp)*D2(ell,mpp,m) for mpp in range(-ell,ell+1)]) ) > Tolerance
    for ell in range(2,9) for mp in range(-ell,ell+1) for m in range(-ell,ell+1)]
if(sum(Differences)==0) :
    print("There are no differences greater than {0}.".format(Tolerance))
else :
    print("The following are (ell,mp,m) for which there are differences greater than {0}:".format(Tolerance))
    print (array([[ell,mp,m] for ell in range(2,9) for mp in range(-ell,ell+1) for m in range(-ell,ell+1)])[array(Differences)]).tolist()
```

### Appendix A 4: Interpolation

It is important to be able to smoothly interpolate rotations.  Here, we construct a series of 5 rotors, along with their respective time steps, and interpolate them to a finer series.  Here, we plot the north pole as rotated by the various rotors.  The original 5 points are plotted as large green circles, and the interpolated data is plotted as small red circles.  We see that the interpolated data is very nicely smoothed.

```python
# Define the basis vectors
x = Quaternions.Quaternion(0,1,0,0)
y = Quaternions.Quaternion(0,0,1,0)
z = Quaternions.Quaternion(0,0,0,1)

# Define some sparse input data to be interpolated
qIn = [Quaternions.exp((0./2.0)*(x)).normalized(),
    Quaternions.exp((math.pi/2./2.0)*(x)).normalized(),
    Quaternions.exp((math.pi/2./2.0)*(x+y).normalized()).normalized(),
    Quaternions.exp((math.pi/2./2.0)*(y)).normalized(),
    Quaternions.exp((8*math.pi/20./2.0)*(x+y).normalized()).normalized()
]
tIn = [-10.0, 5.0, 10.0, 15.0, 20.0]

# Interpolate the data to much finer steps
NSteps = 80
tOut = linspace(tIn[0], tIn[-1], NSteps)
qOut = Quaternions.Squad(qIn, tIn, tOut)

# Visualize in 3-D
vartheta = linspace(0, pi, 100)
varphi = linspace(0, 2 * pi, 100)
Sphere_x = outer(sin(vartheta), cos(varphi))
Sphere_y = outer(sin(vartheta), sin(varphi))
Sphere_z = outer(cos(vartheta), ones_like(varphi))
QuatXYZIn  = array([(R*z*R.conjugate()).vec() for R in qIn])
QuatXYZOut = array([(R*z*R.conjugate()).vec() for R in qOut])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Surface = ax.plot_surface(0.95*Sphere_x, 0.95*Sphere_y, 0.95*Sphere_z,
                          rstride=5, cstride=5, color='#fff46f', linewidth=0.5, antialiased=False)
Dots1 = ax.plot(QuatXYZIn[:,0], QuatXYZIn[:,1], QuatXYZIn[:,2], 'go', ms=9)
Dots2 = ax.plot(QuatXYZOut[:,0], QuatXYZOut[:,1], QuatXYZOut[:,2], 'ro', ms=3)
```
